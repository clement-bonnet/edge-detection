# Report Experiments

```python
import matplotlib.pyplot as plt
import numpy as np
import sobel
```

```python
lena = sobel.SobelDetector("img/lena.jpg")
sobel1 = sobel.SobelDetector("img/7.1.03.tiff")
sobel2 = sobel.SobelDetector("img/5.2.08.tiff")
sobel3 = sobel.SobelDetector("img/5.1.12.tiff")
```

# Lena

```python
lena.show_img(cmap="gray")
# plt.savefig("lena.png", bbox_inches="tight");
```

```python
lena.show_edges(cmap="gray")
# plt.savefig("lena_gradients.png", bbox_inches="tight");
```

```python
plt.hist(lena.edges.flatten(), density=True, bins=100);
plt.xlabel("Gradient magnitude");
plt.ylabel("Density");
plt.title("Histogram of gradient magnitudes");
# plt.savefig("lena_hist.png", dpi=300, bbox_inches="tight");
```

# Quantile Edge Detection

```python
plt.hist(sobel1.edges.flatten(), density=True, bins=100);
plt.vlines(np.quantile(sobel1.edges.flatten(),0.9), ymin=0, ymax=12, colors="red");
plt.xlabel("Gradient magnitude");
plt.ylabel("Density");
plt.title("Histogram of gradient magnitudes");
# plt.savefig("hist1.png", dpi=300, bbox_inches="tight");
```

```python
plt.hist(sobel2.edges.flatten(), density=True, bins=100);
plt.vlines(np.quantile(sobel2.edges.flatten(),0.9), ymin=0, ymax=20, colors="red");
plt.xlabel("Gradient magnitude");
plt.ylabel("Density");
plt.title("Histogram of gradient magnitudes");
# plt.savefig("hist2.png", dpi=300, bbox_inches="tight");
```

```python
plt.hist(sobel3.edges.flatten(), density=True, bins=100);
plt.vlines(np.quantile(sobel3.edges.flatten(),0.9), ymin=0, ymax=43, colors="red");
plt.xlabel("Gradient magnitude");
plt.ylabel("Density");
plt.title("Histogram of gradient magnitudes");
# plt.savefig("hist3.png", dpi=300, bbox_inches="tight");
```
